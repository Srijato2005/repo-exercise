```python
ratings_with_name = df3.merge(df1,on='ISBN')
```

```python
ratings_with_name['Book-Title']=ratings_with_name['Book-Title'].astype(str)
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings'] >=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(df1,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author_x','num_ratings','avg_rating','Image-URL-L_x']]
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
educated=x[x].index # where user has rates 250 
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(educated)]
filtered_rating  
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_score=cosine_similarity(pt)
sim_score
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_books=sorted(list(enumerate(sim_score[index])),key=lambda x:x[1],reverse=True)[1:10]
    for i in similar_books:
        print( pt.index[i[0]])


```
