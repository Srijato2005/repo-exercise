```python
end_date= '2025-10-27' 
start= pd.to_datetime(end_date)-pd.DateOffset(365*8)  
df= yf.download(tickers=list, start=start, end=end_date).stack()
df


```

```python
df.columns= df.columns.str.lower()
```

```python
stock=df.xs('AAPL',level='Ticker')['close'].dropna()
stock
```

```python
data= stock.values.reshape(-1,1)
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale= MinMaxScaler() 
scaled_data= scale.fit_transform(data) 
scaled_data
```

```python
train_data_len= int(len(scaled_data)*0.8) 
train_data= scaled_data[0:train_data_len,:] 
test_data = scaled_data[train_data_len - 60:, :]
```

```python
def create_Windows(data,window=60) :
    x,y=[],[] 
    for i in range(len(data)- window ): 
        x.append(data[i:i+window]) 
        y.append(data[i+window]) 
    return np.array(x) ,np.array(y)
    
```

```python
trainx,trainy= create_Windows(train_data,60) 
testx,testy= create_Windows(test_data,60) 
```

```python
trainx.shape
```

```python
trainx = np.reshape(trainx, (trainx.shape[0], trainx.shape[1],1) ) 
testx = np.reshape(testx, (testx.shape[0], testx.shape[1],1)  )
```

```python
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(60, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2)) #reduces too much dependency on a layer of data 
model.add(Dense(units=1)) 
```

```python
model.compile(optimizer='adam', loss='mean_squared_error') 
```

```python
model.fit( trainx, trainy ,  epochs=100, validation_data=(testx,testy),batch_size= 32) 
```

```python
predictions= model.predict(testx) 

```

```python
predictions= scale.inverse_transform(predictions)
```

```python
actual= scale.inverse_transform(testy.reshape(-1,1) )
                                
```

```python
plt.figure(figsize=(16, 8))
plt.title('Stock Price Prediction (AAPL)')
plt.xlabel('Time (Days in Test Set)')
plt.ylabel('Stock Price USD')
plt.plot(actual, color='blue', label='Actual Price')
plt.plot(predictions, color='red', linestyle='--', label='Predicted Price')
plt.legend()
```
