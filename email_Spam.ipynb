```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
df.drop(['Unnamed: 3','Unnamed: 2'],axis="columns")
```

```python
 from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test= train_test_split(df.Message,df.spam,test_size=0.3)
```

```python
x_train.astype(str)
from sklearn.feature_extraction.text import CountVectorizer
y_train_clean = y_train.fillna('').astype(str)
x_train_clean = x_train.fillna('').astype(str)
v = CountVectorizer()
x_train_count = v.fit_transform(x_train_clean)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
y_train_clean = y_train.fillna('').astype(str)
x_train_clean = x_train.fillna('').astype(str)
model.fit(x_train_count,y_train_clean)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```
