```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
def chunk_data(doc,chunk_size=1000,  chunk_overlap=200):
    splitter = RecursiveCharacterTextSplitter(
                            chunk_size=chunk_size,
                            chunk_overlap=chunk_overlap
                               )
    document= splitter.split_documents(doc)

    return document
```

```python
documents= chunk_data(doc)
documents
```

```python
from re import search
retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={"k": 2})

```

```python
from langchain_core.prompts import PromptTemplate
```

```python
llm= ChatGoogleGenerativeAI(model="gemini-2.0-flash-001")
prompt= PromptTemplate(
    template="You are a helpful ai assistant, provide nswers only from the provided data and if not present clearly mention that u dont know {context} question:-{question}",
    input_variables=["context","question"]
)
```

```python
question= "What are few of the reviews of the product "
context= retriever.invoke(question)

```

```python
from langchain.schema.runnable import RunnableBranch,RunnablePassthrough,RunnableSequence,RunnableParallel,RunnableLambda
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser= StrOutputParser()
```

```python
def format(context):
  context_text= "\n\n".join(doc.page_content for doc in context)
  return context_text
```

```python
parallel_chain= RunnableParallel({
    "context": retriever| RunnableLambda(format),
    "question":RunnablePassthrough()
})
parallel_chain.invoke("What is the iphone 16 features  ?")
```

```python
main_chain= parallel_chain| prompt|llm|parser
```

```python
main_chain.invoke("summarize the file ")
```
