FEATURE EXTRACTION AND ENGINEERING

```python
df['AVERAGE_BILL_AMT']= np.mean(df[['BILL_AMT1',	'BILL_AMT2',	'BILL_AMT3',	'BILL_AMT4',	'BILL_AMT5',	'BILL_AMT6']], axis=1)
df
```

```python
df['AVERAGE_PAY_AMT']= np.mean(df[['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']], axis=1)
df
```

```python
df['DELAY']=np.mean(df[['PAY_0',	'PAY_2'	,'PAY_3',	'PAY_4','PAY_5','PAY_6']], axis=1)
df
```

```python
df['UTILISATION']=df['AVERAGE_BILL_AMT']/df['LIMIT_BAL']
df
```

```python
df['PAYMENT_RATIO']=df['AVERAGE_PAY_AMT']/df['AVERAGE_BILL_AMT']
df
```

SCALING THE DATA

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
features_to_scale=['LIMIT_BAL','AGE','AVERAGE_BILL_AMT','AVERAGE_PAY_AMT','DELAY','UTILISATION','PAYMENT_RATIO']
df.replace([-np.inf,np.inf],np.nan,inplace=True)
df_scaled=df.dropna(subset=features_to_scale).copy()
df_scaled[features_to_scale]=scaler.fit_transform(df_scaled[features_to_scale])
df_scaled
```

CLUSTERING OF DATA

```python
from sklearn.cluster import KMeans
```

check how many clusters to be provided by ELBOW METHOD

```python
sse=[]
k_range= range(1,10)
for k in range(1,10):
  km=KMeans(n_clusters=k)
  km.fit(df_scaled)
  sse.append(km.inertia_)







```

```python
import matplotlib.pyplot as plt
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_range, sse)
plt.show()
```

taking K=3

```python
km=KMeans(n_clusters=3)
features_to_use=['LIMIT_BAL','AGE','AVERAGE_BILL_AMT','AVERAGE_PAY_AMT','DELAY','UTILISATION','PAYMENT_RATIO','SEX','MARRIAGE','EDUCATION']

y_pred=km.fit_predict(df_scaled[features_to_use])
y_pred
```

```python
df_scaled['CLUSTERS']=y_pred
df_scaled
```

TRIAL 1 FOR CLIPPING OF UTILISATION DATA FOR OUTLIERS

```python
df['UTILISATION'] = pd.to_numeric(df['UTILISATION'], errors='coerce')
df['UTILISATION'] = df['UTILISATION'].clip(
    lower=df['UTILISATION'].quantile(0.01),
    upper=df['UTILISATION'].quantile(0.99)
)
df['UTILISATION']
```
